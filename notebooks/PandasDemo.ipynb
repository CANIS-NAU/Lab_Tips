### Merging Dataframes

```python
df = pd.DataFrame(np.random.randn(10, 4))
df
```

```python
# Break the dataframe into different parts
pieces = [df[:3], df[3:7], df[7:]]

# Merge the parts
pd.concat(pieces)
```

### SQL Style Joins

```python
# left = pd.DataFrame({'key': ['foo', 'foo'], 'lval': [1, 2]})
left = pd.DataFrame({'key': ['foo', 'bar'], 'lval': [1, 2]})
left
```

```python
# right = pd.DataFrame({'key': ['foo', 'foo'], 'rval': [4, 5]})
right = pd.DataFrame({'key': ['foo', 'bar'], 'rval': [4, 5]})
right
```

```python
pd.merge(left, right, on='key')
```

### Grouping data in a dataframe

```python
df = pd.DataFrame({'A': ['foo', 'bar', 'foo', 'bar',
                          'foo', 'bar', 'foo', 'foo'],
                    'B': ['one', 'one', 'two', 'three',
                             'two', 'two', 'one', 'three'],
                    'C': np.random.randn(8),
                    'D': np.random.randn(8)})
df
```

```python
df.groupby('A').sum()
```

```python
df.groupby(['A', 'B']).sum()
```

### Plot the cumsum of a random dataframe

```python
# Create a one column dataframe i.e. Series
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2000', periods=1000))
ts = ts.cumsum()
ts.plot();
```

```python
# Plot a DataFrame with multiple columns
df = pd.DataFrame(np.random.randn(1000, 4), index=ts.index, columns=['A', 'B', 'C', 'D'])
df = df.cumsum()
df.plot();
```
